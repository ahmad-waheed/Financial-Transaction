```python

mapping = {
    "Sale of software maintenance contracts to existing customers": {
        "category": "Revenue",
        "subcategories": [
            "Maintenance contract sales revenue",
            "Revenue from software support contracts",
            "Service agreement income from existing clients"
        ]
    },
    "Licensing fees for proprietary software used in industrial automation": {
        "category": "Revenue",
        "subcategories": [
            "Software licensing revenue from industrial clients",
            "Proprietary software licensing income",
            "Fees from industrial software usage rights"
        ]
    },
    "Subscription revenue from a cloud-based project management platform": {
        "category": "Revenue",
        "subcategories": [
            "Cloud-based project management subscription revenue",
            "Subscription income from project management services",
            "Recurring revenue from cloud platform subscriptions"
        ]
    },
    "Revenue from training workshops on software development methodologies": {
        "category": "Revenue",
        "subcategories": [
            "Training workshop revenue for software methodologies",
            "Income from software development training sessions",
            "Workshop fee revenue for software practices"
        ]
    },
    "Royalty income from the sale of software products to international distributors": {
        "category": "Revenue",
        "subcategories": [
            "Royalty revenue from international software sales",
            "Income from software product distribution royalties",
            "Royalties earned from global software distribution"
        ]
    },
    "Subscription renewal for cloud-based software service": {
        "category": "Revenue",
        "subcategories": [
            "Software as a Service (SaaS) subscription renewal",
            "Cloud service subscription revenue",
            "Recurring revenue from software subscriptions"
        ]
    },
    "Sale of software licenses to corporate clients": {
        "category": "Revenue",
        "subcategories": [
            "Software license sales",
            "Enterprise software sales",
            "Perpetual software license revenue"
        ]
    },
    "Consultation fee for implementing customized software solutions": {
        "category": "Revenue",
        "subcategories": [
            "Software implementation services",
            "Custom software development consulting",
            "Professional services revenue"
        ]
    },
    "Advertisement revenue from online platform services": {
        "category": "Revenue",
        "subcategories": [
            "Online advertising revenue",
            "Digital advertising sales",
            "Platform monetization through ads"
        ]
    },
    "Licensing fee for intellectual property rights on software products": {
        "category": "Revenue",
        "subcategories": [
            "Intellectual property licensing revenue",
            "Software IP licensing fees",
            "Royalties from software licensing"
        ]
    },
    "Payment for software development outsourcing services": {
        "category": "Expense",
        "subcategories": [
            "Outsourced software development expenses",
            "External development costs",
            "Third-party development fees"
        ]
    },
    "Purchase of computer equipment for software development": {
        "category": "Expense",
        "subcategories": [
            "Hardware procurement for software development",
            "Computer equipment purchases",
            "Technology infrastructure expenses"
        ]
    },
    "Cloud service subscription fee for hosting software applications": {
        "category": "Expense",
        "subcategories": [
            "Cloud hosting expenses",
            "Infrastructure as a Service (IaaS) subscription fees",
            "Hosting costs for software applications"
        ]
    },
    "Marketing expenses for promoting software products": {
        "category": "Expense",
        "subcategories": [
            "Software marketing costs",
            "Advertising and promotion expenses",
            "Digital marketing expenditures"
        ]
    },
    "Employee salaries for software engineering team": {
        "category": "Expense",
        "subcategories": [
            "Software development team salaries",
            "Engineering payroll expenses",
            "Compensation for software engineers"
        ]
    },
    "Research and development expenses for enhancing artificial intelligence algorithms": {
        "category": "Expense",
        "subcategories": [
            "AI algorithm enhancement R&D expenses",
            "Investment in AI research and development",
            "Costs associated with AI algorithm improvements"
        ]
    },
    "Purchase of software development tools and licenses from third-party vendors": {
        "category": "Expense",
        "subcategories": [
            "Third-party software tools and licenses procurement costs",
            "Expenses for acquiring development software tools",
            "Licensing fees for third-party development software"
        ]
    },
    "Hosting fees for a website providing software-as-a-service (SaaS) solutions": {
        "category": "Expense",
        "subcategories": [
            "SaaS website hosting expenses",
            "Cloud hosting fees for SaaS solutions",
            "Costs for hosting software services online"
        ]
    },
    "Advertising expenses for promoting a new mobile application": {
        "category": "Expense",
        "subcategories": [
            "Mobile app advertising costs",
            "Marketing expenditures for app promotion",
            "Promotion expenses for new mobile software"
        ]
    },
    "Travel expenses for attending technology conferences to showcase software products": {
        "category": "Expense",
        "subcategories": [
            "Conference travel expenses for software showcase",
            "Costs related to technology event participation",
            "Expenses for showcasing software at industry conferences"
        ]
    },
    "Acquisition of intellectual property rights for a new software product": {
        "category": "Asset",
        "subcategories": [
            "Intellectual property acquisition cost",
            "Purchase of software patents",
            "Investment in software IP assets"
        ]
    },
    "Investment in research and development for software innovation": {
        "category": "Asset",
        "subcategories": [
            "R&D investment in software development",
            "Research expenses for software innovation",
            "Development of proprietary software technology"
        ]
    },
    "Purchase of office space for establishing software development center": {
        "category": "Asset",
        "subcategories": [
            "Real estate acquisition for software development center",
            "Office property investment for software operations",
            "Asset acquisition for software facility"
        ]
    },
    "Acquisition of patents related to software algorithms": {
        "category": "Asset",
        "subcategories": [
            "Patent acquisition for software algorithms",
            "Investment in software algorithm patents",
            "Purchase of intellectual property rights for algorithmic inventions"
        ]
    },
    "Purchase of software tools and licenses for development purposes": {
        "category": "Asset",
        "subcategories": [
            "Software tool acquisition for development",
            "Development software license purchases",
            "Investment in software development infrastructure"
        ]
    },
    "Investment in blockchain technology for developing decentralized applications": {
        "category": "Asset",
        "subcategories": [
            "Blockchain technology investment for DApps development",
            "Expenditure on blockchain infrastructure for decentralized apps",
            "Costs associated with blockchain development initiatives"
        ]
    },
    "Acquisition of software patents to protect intellectual property rights": {
        "category": "Asset",
        "subcategories": [
            "Patent acquisition costs for software IP protection",
            "Expenses for securing software patent rights",
            "Investment in intellectual property through patent acquisitions"
        ]
    },
    "Purchase of server infrastructure for hosting software applications": {
        "category": "Asset",
        "subcategories": [
            "Server infrastructure acquisition for software hosting",
            "Costs for purchasing servers for software deployment",
            "Investment in server hardware for application hosting"
        ]
    },
    "Capitalization of software development costs for internally developed software": {
        "category": "Asset",
        "subcategories": [
            "Internal software development costs capitalized",
            "Capitalized costs for in-house software projects",
            "Expenditure recorded as assets for internally developed software"
        ]
    },
    "Investment in cybersecurity software to enhance data protection measures": {
        "category": "Asset",
        "subcategories": [
            "Cybersecurity software investment for data protection",
            "Expenses for enhancing data security with software solutions",
            "Costs associated with cybersecurity software implementation"
        ]
    },
    "Loan obtained for funding software development projects": {
        "category": "Liability",
        "subcategories": [
            "Software development project financing loan",
            "Debt financing for software development",
            "Loan liability for software projects"
        ]
    },
    "Accrued expenses for software maintenance contracts": {
        "category": "Liability",
        "subcategories": [
            "Accrued software maintenance costs",
            "Liabilities for ongoing software support",
            "Accruals for software maintenance services"
        ]
    },
    "Deferred revenue from pre-paid software subscriptions": {
        "category": "Liability",
        "subcategories": [
            "Deferred revenue from prepaid software subscriptions",
            "Unearned revenue from software subscriptions",
            "Liabilities from advance payments for software services"
        ]
    },
    "Lease liability for office space rented for software operations": {
        "category": "Liability",
        "subcategories": [
            "Office lease liability for software operations",
            "Lease obligation for software facility",
            "Rental liability for software office space"
        ]
    },
    "Accounts payable to software vendors for purchased licenses": {
        "category": "Liability",
        "subcategories": [
            "Accounts payable for software licenses",
            "Payables to software vendors",
            "Vendor liabilities for software purchases"
        ]
    },
    "Accrued liabilities for software license agreements yet to be invoiced": {
        "category": "Liability",
        "subcategories": [
            "Accrued software license agreement liabilities",
            "Liabilities for un-invoiced software licenses",
            "Expenses accrued for pending software license agreements"
        ]
    },
    "Deferred revenue from pre-sales of upcoming software releases": {
        "category": "Liability",
        "subcategories": [
            "Deferred revenue from pre-sales of software releases",
            "Unearned revenue from software pre-sales",
            "Liabilities from advance payments for future software releases"
        ]
    },
    "Loan obligations for financing the acquisition of software companies": {
        "category": "Liability",
        "subcategories": [
            "Loan liabilities for software company acquisitions",
            "Debt financing obligations for software acquisitions",
            "Expenses associated with loan repayments for software acquisitions"
        ]
    },
    "Lease obligations for equipment used in software testing facilities": {
        "category": "Liability",
        "subcategories": [
            "Equipment lease liabilities for software testing",
            "Lease obligations for testing facility equipment",
            "Costs associated with leased testing equipment"
        ]
    },
    "Accounts payable to software development contractors for project milestones": {
        "category": "Liability",
        "subcategories": [
            "Payables to software development contractors",
            "Contractor payment obligations for project milestones",
            "Expenses accrued for contractor services in software development"
        ]
    },
    "Issuance of new shares to raise capital for software expansion": {
        "category": "Equity",
        "subcategories": [
            "Equity capital raised through share issuance",
            "Stock issuance for software expansion",
            "Share capital raised for software growth"
        ]
    },
    "Retained earnings reinvested into research and development": {
        "category": "Equity",
        "subcategories": [
            "Reinvestment of retained earnings into R&D",
            "Equity reinvestment for innovation and development",
            "Retained earnings allocated to software improvement"
        ]
    },
    "Stock options granted to employees as part of compensation": {
        "category": "Equity",
        "subcategories": [
            "Equity-based compensation through stock options",
            "Employee stock option grants",
            "Equity incentives for staff through options"
        ]
    },
    "Dividends declared to shareholders from profits generated by software sales": {
        "category": "Equity",
        "subcategories": [
            "Dividend distribution from software profits",
            "Equity dividends paid to shareholders",
            "Shareholder returns from software sales"
        ]
    },
    "Contribution of capital by founders for initial software development": {
        "category": "Equity",
        "subcategories": [
            "Founders' capital contribution for software startup",
            "Equity investment by founders in software venture",
            "Initial capital injection for software development"
        ]
    },
    "Capital injection from venture capital firms for scaling software operations": {
        "category": "Equity",
        "subcategories": [
            "Venture capital investment for software scaling",
            "Funds raised from venture capitalists for software expansion",
            "Equity financing for software growth from VC firms"
        ]
    },
    "Distribution of bonus shares to employees as part of performance incentives": {
        "category": "Equity",
        "subcategories": [
            "Employee bonus share distribution",
            "Equity incentives through bonus share allocation",
            "Distribution of shares as performance rewards"
        ]
    },
    "Declaration of stock dividends to shareholders based on company profitability": {
        "category": "Equity",
        "subcategories": [
            "Stock dividend declaration based on company profits",
            "Dividend distribution to shareholders in stock form",
            "Equity dividends declared to shareholders"
        ]
    },
    "Retention of earnings for reinvestment in expanding software product lines": {
        "category": "Equity",
        "subcategories": [
            "Retained earnings reinvested in software product expansion",
            "Reinvestment of profits for software product line growth",
            "Equity retained for reinvestment in software ventures"
        ]
    },
    "Allocation of equity stakes to co-founders based on their contributions to software development": {
        "category": "Equity",
        "subcategories": [
            "Equity stake allocation to co-founders",
            "Distribution of equity based on co-founders' contributions",
            "Equity allocation for founding team members"
        ]
    }
}
fake = Faker()
def adjust_amount(description, amount):

    #Adjusts the amount based on the description and subcategory.

    # Increase amount for revenue
    if mapping[description]["category"] == "Revenue":
        amount *= random.uniform(1.1, 1.5)
    # Decrease amount for expenses
    elif mapping[description]["category"] == "Expense":
        amount *= random.uniform(0.5, 0.9)

    return round(amount, 2)



def generate_data(mapping, num_entries=2500):
    data = []
    vendors_consumers = []
    for _ in range(num_entries):
        desc, info = random.choice(list(mapping.items()))
        category = info["category"]
        subcategory = random.choice(info["subcategories"])
        amount = round(random.uniform(1000, 1000000), 2)
        amount = adjust_amount(desc, amount)
        data.append([desc, amount, category, subcategory])
        if len(vendors_consumers) < num_entries // 5:
            vendor_consumer = fake.company() if random.random() < 0.7 else random.choice(["Apple Inc.", "Microsoft Corporation", "Amazon.com, Inc.", "Google LLC"])
            vendors_consumers.extend([vendor_consumer] * random.randint(2, 4))  # Repeat 2-4 times
    vendors_consumers.extend([fake.company() for _ in range(num_entries - len(vendors_consumers))])  # Extend the list to match DataFrame length
    random.shuffle(vendors_consumers)  # Shuffle to randomize the order
    return data, vendors_consumers

# Generate data
data, vendors_consumers = generate_data(mapping)

# Convert to DataFrame
columns = ['Description', 'Amount', 'Category', 'Subcategory']
df = pd.DataFrame(data, columns=columns)

# Add other columns
df['Date'] = pd.Timestamp.now().strftime('%Y-%m-%d')
df['Company ID'] = [random.randint(1000, 9999) for _ in range(df.shape[0])]
df['Industry'] = 'Software and Tech'

# Assign vendors/consumers from the generated list
df['vendor_or_consumer'] = vendors_consumers[:len(df)]

df['Payment Method'] = [random.choice(['Check', 'Cash', 'Bank Transfer','Card']) for _ in range(df.shape[0])]
df['Currency'] = 'USD'
df['Geographical Location'] = 'USA'
df['Account Number'] = [random.randint(10000000, 99999999) for _ in range(df.shape[0])]
df['Reference Number'] = [random.randint(1000000000, 9999999999) for _ in range(df.shape[0])]
df['Approval Status'] = [random.choice(['Approved', 'Pending', 'Rejected']) for _ in range(df.shape[0])]
df['Customer Type'] = df['Description'].apply(lambda x: 'New Customer' if 'new' in x.lower() else 'Existing Customer')

# Save to CSV
df.to_csv('Software_industry.csv', index=False)

# Display first few rows
print(df.head())
```
